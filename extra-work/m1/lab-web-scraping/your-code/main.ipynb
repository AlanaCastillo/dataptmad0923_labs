````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
response =requests.get(url)
````

````python
zip_codes = response.content
len(zip_codes)
````

````python
parsed_codes = bs4.BeautifulSoup(zip_codes, 'html.parser')
parsed_codes
````
